```python
import pandas as pd 
import numpy as np
import io
import matplotlib.pyplot as plt
```

Dados do zap Imóveis e imóvel web

```python
file="Imoveis_JACAREI_24022023.txt"

from google.colab import files
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded[file]),sep=";")
```

```python
df.head(10)
```

```python
df['Preco_m2']=df['valor']/df['m2']
```

```python
df.describe()
```

```python
#análise Apto 72 m2
df.describe()["Preco_m2"]["mean"]*72
```

Preço do aluguel proposto pelo nivaldo (1700) para o apto do condomínio green ville (72 m2) é um pouco menor do que a média para o local

```python
preco_nivaldo=1700/72
plt.hist(df['Preco_m2'])
plt.vlines(preco_nivaldo,0,4)
plt.legend()
plt.show()
```

Preço aluguel estimado pelo Nivaldo é compatível com o observado na região do outro lado da ponte/Jardim dindinha com dados extraídos do Imovelweb e zap imóveis.

A média é um pouco distorcida por alguns aptos, provavelmente de alto padrão, que apresentam uma distribuição mais a direita.

O Apto do Nivaldo possui móveis planejados e armários e em geral os imóveis na região aparentavam não ter nada. Aparentemente o preço é razoável.

```python
#remover os aptos de alto padrão
df=df[df["Preco_m2"]<30]
df.describe()
```

```python
#análise Apto 72 m2
df.describe()["Preco_m2"]["mean"]*72
```

```python
preco_nivaldo=1700/72
plt.hist(df['Preco_m2'])
plt.vlines(preco_nivaldo,0,4)
plt.legend()
plt.show()
```
